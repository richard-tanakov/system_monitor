```python
import psutil
from rich import print 

def acquisition_cpu():
    cpu_usage = psutil.cpu_percent(interval=1)
    cpu_per_core = psutil.cpu_percent(interval=1, percpu=True)

    return {
        "physical": psutil.cpu_count(logical=False),
        "logical": psutil.cpu_count(logical=True),
        "total_usage": cpu_usage,
        "per_core_usage": cpu_per_core
    }




def memory():
    mem = psutil.virtual_memory()
    swap = psutil.swap_memory()

    return {
        "total_memory_mb": mem.total // (1024 ** 2),
        "used_memory_mb": mem.used // (1024 ** 2),
        "free_memory_mb": mem.available // (1024 ** 2),
        "memory_usage_percent": mem.percent,
        "total_swap_mb": swap.total // (1024 ** 2),
        "used_swap_mb": swap.used // (1024 ** 2),
    }
```

```python
cpu = acquisition_cpu()
mem = memory()

print("[bold cyan]CPU:[/bold cyan]", cpu)
print("[bold green]Memory:[/bold green]", mem)
```
